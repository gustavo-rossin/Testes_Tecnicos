### Parte 1 - ETL Dos Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
#trazendo a tabela para o Python

#utilizei encoding latin1 por conta do arquivo não conseguir ler utf-8

df = pd.read_csv('case_atendimentos.csv', encoding='latin1', delimiter=';')
df.head()
```

```python
#Processo de análise de dados: pt 1 - checando se há duplicatas
df.duplicated().sum()
```

```python
#Processo de análise de dados: pt 2 - informação de cada coluna
df.info()
```

```python
# Processo de análise de dados: pt 3 - verificando se a base tem dados com NaN
df.isna().sum()
```

```python
#Processo de análise de dados: pt 4 - checando os indicadores estatísticos para ver se não há dados com anomalia.
df.describe()

df.info()
#valores de tempo_total_chamada_minutos muito alto. Vou deixar todos os valores com a mediana.
```

```python
#alguns dados com outliers foram encontrados - utilizei a mediana para substituir valores das chamadas e a média para substituir das idades

df['tempo_total_chamada_minutos'] = np.where(df['tempo_total_chamada_minutos'] > 100, 5, df['tempo_total_chamada_minutos'])
df['idade'] = np.where(df['idade'] > 100, 30, df['idade'])

df.describe()
```

```python
print(df.sexo_paciente.value_counts())
```

```python
#modelando dados para montar um gráfico de dispersao sobre sexo do paciente e tempo de chamada

sexo_f_ou_m = df[((df['sexo_paciente'] == 'F') | (df['sexo_paciente'] == 'M')) & (df['tempo_total_chamada_minutos'] > 0)]

sexo_f_ou_m.describe()
```

## Parte 2 - Plotagem dos gráficos

#### Sobre sexo, idade e tempo de ligação

```python
sns.histplot(data = sexo_f_ou_m, x = sexo_f_ou_m['idade'], kde = True, hue = sexo_f_ou_m['sexo_paciente'], multiple='dodge')
sns.set_palette('PiYG')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.title('Quantidade de pessoas pela idade')

#distribuição de pessoas que marcaram o sexo como F ou M. Nota-se que temos a maior quantidade de mulheres e que a maior quantidade de mulheres está entre 30 e 40 anos. Homens na faixa entre20 e 30
```

```python
#linha de regressão levemente positiva: quanto maior a idade da pessoa, mais tempo ela tende a ficar na chamada.
sns.set_palette('PiYG')
sns.regplot(data = sexo_f_ou_m, x = sexo_f_ou_m['tempo_total_chamada_minutos'], y = sexo_f_ou_m['idade'], line_kws=dict(color='k'), label = 'Pessoas', marker='s')

plt.xlabel('Total de minutos na chamada')
plt.ylabel('Idade')
plt.title('Pessoas por tempo no telefone')
plt.legend()
plt.show()
```

```python
sexo_f_ou_m.corr()
```

```python
#definindo a variavel dependende (y) e a independente (X) para encontrar o R² que vai nos mostrar o quanto este modelo de regressão se ajusta aos dados observados.
y = sexo_f_ou_m['idade']
X = sexo_f_ou_m['tempo_total_chamada_minutos']
```

```python
X = sm.add_constant(X)

modelo = sm.OLS(y, X)
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

```python
nota_medico = df.groupby('nps_medico')['id_atendimento'].count().dropna().reset_index().rename(columns = {'id_atendimento': 'contagem'})
nota_medico = nota_medico.sort_values(by='contagem', ascending = False)




sns.set_palette('PiYG')
plt.figure(figsize=(8,6))
ax = sns.barplot(data = nota_medico, x = 'nps_medico', y = 'contagem')
plt.xlabel('Nota do Atendimento')
plt.ylabel('Contagem')
plt.title('Nota atendimento pela quantidade')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
    height + 3,
    '{:1.0f}'.format((height)),
    ha='center')

ax
```

```python
total_notas = nota_medico.contagem.sum()

(3184+349+174+50) / total_notas
```

```python
# TOP 5 especialidades atendidas

nome_especialidade = df.groupby('nome_especialidade')['nps_medico'].agg([np.mean, 'count']).dropna().reset_index()

nome_especialidade_filtrada = nome_especialidade.sort_values(by = 'count', ascending= False).head(5)

nome_especialidade_filtrada['mean'] = round(nota_especialidade_filtrada['mean'], 2)


sns.set_palette('PiYG')
plt.figure(figsize = (8, 4))
ax = sns.barplot(data = nome_especialidade_filtrada, y = 'count', x = 'mean', hue = 'nome_especialidade', dodge = False)
plt.xlabel('Média de Notas')
plt.ylabel('Contagem de Atendimentos')
plt.title('TOP 5 especialidades atendidas')


for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 3, '{:1.0f}'.format((height)), ha='center')

ax
```

```python
# TOP 4 PRINCIPAIS ATENDIMENTOS REALIZADOS

tipos_atendimento = df.objetivo_atendimento.value_counts()
tipos_atendimento_contagem = tipos_atendimento.reset_index().rename(columns = {'index': 'objetivo_atendimento', 'objetivo_atendimento': 'contagem'})
pizza_filtrada = tipos_atendimento_contagem.head(4)
sns.set_palette('PiYG')
plt.pie(data = pizza_filtrada, x = 'contagem', labels = 'objetivo_atendimento', autopct='%.0f%%', textprops={'size': 'smaller'}, radius=0.8, startangle = 0)
plt.title('Principais Objetivos de Atendimento')
plt.figure(figsize = (10, 5))
```

```python
# Plot para identificar qual o estado que tem mais pacientes usando o serviço da Conexa.

uf_pacientes = df.groupby('uf_paciente')['id_paciente'].nunique().dropna().reset_index().rename(columns = {'id_paciente': 'contagem'})

uf_pacientes = uf_pacientes.sort_values(by = 'contagem', ascending = False)

ax = sns.barplot(data = uf_pacientes, x = 'uf_paciente', y = 'contagem')
plt.xlabel('Origem dos Pacientes')
plt.ylabel('Quantidade')
plt.figure(figsize = (8,4))

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 3, '{:1.0f}'.format((height)), ha='center')

ax
```
